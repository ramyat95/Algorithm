### Sorting Algorithm
- Arranging the list of elements according to comparision operator
- It can be 
    - Simple sort (Bubble sort, selection sort , Insertion sort)
    - Complex sort (Quick sort, Merge sort, Timesort)

## Bubble sort
- Compare and swapping two elements like small soap bubbles and hence the name given as bubble sort.


```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if (arr[j]>arr[j+1]):
                arr[j] , arr[j+1] = arr[j+1], arr[j]
                
arr = [7,4,5,2]
bubble_sort(arr)
for i in range(len(arr)):
    print(arr[i])
```

## Selection sort
- The Selection sort algorithm is based on the idea of finding the minimum or maximum element in an unsorted array and then putting it in its correct position in a sorted array.


```python
                
a1 = [10,14,27,33,35,19,42,44]
n = len(a1)
for i in range(n):
    low = i
    for j in range(i+1,n):
        if (a1[i]>a1[j]):
            low=j
    a1[i],a1[low]=a1[low],a1[i]

for i in range(len(a1)):
    print(a1[i])        
        
```

## Insertion sort
- Insertion sort is a simple sorting algorithm that builds the final sorted array (or list) one item at a time.
- An insertion sort is less efficient on large lists than more advanced algorithms such as quicksort, heapsort, or merge sort. 
- A simple example of insertion sort is when people manually sort cards in their hands so they have a clearer picture of how to play their cards. 
- Insertion sorts provide several advantages:
    - Simple implementation
    - Efficient for very small data sets, can sort a list as it receives it
    - Memory efficient
    

```python
def insertion_sort(a2):
    for i in range(len(a2)):
        for j in range(0,i+1):
            if (a2[i]<a2[j]):
                a2[i],a2[j]=a2[j],a2[i]
a2 = [3,7,5,6,9]
insertion_sort(a2)
for i in range(len(a2)):
    print(a2[i])
            
```

## Quick sort
- Quick Sort is a type of divide and conquer algorithm. 
- It picks an element (numeric or alphanumeric) as a pivot and then partitions the array around the pivot. 
- There are many different methods for a quick sort to pick a pivot
        - Pick the first element as pivot.
        - Pick the last element as pivot
        - Pick a random element as pivot.
        - Pick the median as pivot.
        


```python
def partition(array, low, high):
    i = (low - 1)
    pivot = array[high]
    for j in range(low, high):
        if array[j] <= pivot:
            i += 1
            array[i], array[j] = array[j], array[i]
    array[i + 1], array[high] = array[high], array[i + 1]
    return (i + 1)

def quick_sort(array, low, high):
    if low < high:
        partition_index  = partition(array, low, high)
        quick_sort(array, low, partition_index  - 1)
        quick_sort(array, partition_index  + 1, high)
        
array = [1, 0, 4, 2, 3, 5]
low = 0
high = len(array)-1
quick_sort(array, low, high)
print(array)    
```

## Merge sort
- Merge Sort divides an input array (numeric or alpha numeric) into two halves and then merges the two sorted halves. A merge() function is used for merging the two halves.

## Timsort
- TimSort is a sorting algorithm that is a sort based on Insertion Sort and Merge Sort.  
- Timsort first analyses a list that it is trying to sort and then chooses the best sorting approach based on the analysis of the list.
